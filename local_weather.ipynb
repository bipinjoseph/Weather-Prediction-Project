```python
def create_predictions(predictors,core_weather,reg):
     train =core_weather.loc[:"2020-12-31"]
     test = core_weather.loc["2021-01-01":]
     reg.fit(train[predictors],train["target"])
     predictions= reg.predict(test[predictors])
     error=mean_absolute_error(test["target"],predictions)
     combined = pd.concat([test["target"],pd.Series(predictions,index=test.index)],axis=1)
     combined.columns= ["actual","predictions"]
     return error,combined
```

```python
core_weather["month_max"]=core_weather["temo_max"].rolling(30,min_periods=1).mean()
```

```python
core_weather
```

```python
core_weather["temp_min"] = core_weather["temp_min"].replace(0, np.nan)
core_weather["temp_min"] = core_weather["temp_min"].fillna(core_weather["temp_min"].median())
```

```python
core_weather["month_day_max"]=core_weather["month_max"]/core_weather["temo_max"]
```

```python
core_weather["max_min"]=core_weather["temo_max"]/core_weather["temp_min"]
```

```python
core_weather = core_weather.replace([np.inf, -np.inf], np.nan).dropna()
```

```python
core_weather = core_weather.fillna(core_weather.mean())
```

```python
predictors=["precip","temo_max","temp_min","month_max","month_day_max","max_min"]
```

```python
core_weather=core_weather.iloc[30:,:].copy()
```

```python
error , combined= create_predictions(predictors,core_weather,reg)
```
